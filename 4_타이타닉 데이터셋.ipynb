### 3-3. 라벨 인코딩(Label Encoding)

- 문자(Categorical)를 수치(Numerical)로 변환

```python
# 실제 모델에 넣어줄 성별 컬럼만 라벨 인코딩 진행!
df.info()
```

```python
df['Sex'].value_counts()
```

```python
# 매개변수가 male 이면 0을 리턴
# female 이면 1을 리턴하는 함수 작성
def convert_sex(data):
  if data == 'male':
    return 0
  elif data == 'female':
    return 1
```

```python
# 위 함수를 사용하여 Sex 필드 데이터를 수정
df['Sex'] = df['Sex'].apply(convert_sex)
```

```python
df['Sex']
```

```python
# 사이킷런을 이용한 라벨 인코딩
from sklearn.preprocessing import LabelEncoder
```

```python
df['Embarked'].value_counts()
```

```python
le = LabelEncoder()
```

```python
embarked = le.fit_transform(df['Embarked'])
embarked
```

```python
le.classes_
```

### 3-4. 원 핫 인코딩(One Hot Encoding)
- 독립적인 데이터는 별도의 컬럼으로 분리하고 각각 컬럼에 해당 값에만 1, 나머지는 0값으로 갖게 하는 방법
- 예) 머신러닝 알고리즘은 0:'C', 1:'Q', 2:'S', 3:nan 데이터의 관계성을 찾아 'Q+Q = S'라고 학습을 할 수 있음 -> 관계성을 끊어주기 위해 원 핫 인코딩을 사용


```python
# 라벨 인코딩한 걸 일단 추가!
df['Embarked_num'] = embarked
```

```python
pd.get_dummies(df['Embarked_num'])
```
